Preprocessing

가사에서 영어 50% 넘으면 삭제

````python
def drop_if_eng(df): 
    df['one_string'] = df['Lyric'].str.join(" ") 
    mecab = Mecab() 
    for i in df.one_string:
        sents = mecab.pos(i) 
        n = 0 # n: SL(외국어)태그의 숫자
        result = []
        for word in sents:
            if word[1] == 'SL': # SL: Mecab에서 외국어 의미
                n += 1
        if n > (len(sents)//2): # 50% 넘으면 제거 
            df = df[df.one_string != i] 
            #print(len(df)) 
    return df
````

````python
df_saved = drop_if_eng(df_allsongs)
````
