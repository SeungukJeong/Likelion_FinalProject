```python
import pandas as pd 
import os
```

```python
os.getcwd()
```

```python
df_sentiment_0 = pd.read_csv('data_for_mergeLyric/song_sentiment_tag.csv') 
df_sentiment_1 = pd.read_csv('data_for_mergeLyric/songtagged_may30_0.csv')
df_sentiment_2 = pd.read_csv('data_for_mergeLyric/songtagged_may30_1.csv')
df_sentiment_3 = pd.read_csv('data_for_mergeLyric/songtagged_may29_left.csv')

df_songid = pd.read_csv('data_for_mergeLyric/df_allsongs_may30.csv')
```

```python
df_sentiment = pd.concat([df_sentiment_0, df_sentiment_1, df_sentiment_2, df_sentiment_3])
df_sentiment = df_sentiment.reset_index(drop = True) 
df_sentiment = df_sentiment[df_sentiment['Emotion'] != ' ']
df_sentiment = df_sentiment.reset_index(drop = True) 
df_sentiment
```

```python
df_songid.SongID = df_songid.SongID.astype(str)
```

```python
df_sentiment['Title'].nunique() 
df_sentiment.drop_duplicates(subset = ['Title'], inplace = True) 
df_sentiment = df_sentiment.reset_index(drop = True) 
del df_sentiment['Unnamed: 0']
df_sentiment.drop_duplicates()
df_sentiment
```

```python
df_merged = pd.merge(df_sentiment, df_songid[['Title', 'SongID']], on = 'Title', how = 'left')
cols = df_merged.columns.tolist()
cols = ['Title',
        'SongID',
        'Artist',
        'Date',
        'Genre',
        'preprocess_Lyric_ver3',
        'Emotion',
        'Probability',
        '분노혐오',
        '놀람공포',
        '슬픔',
        '행복'
        ]
df_merged = df_merged[cols]
df_merged = df_merged.rename(columns = {'preprocess_Lyric_ver3':'Lyrics'})
df_merged
```

```python
df_merged['Title'].nunique() 
df_merged.drop_duplicates(subset = ['Title'], inplace = True) 
```

```python
df_merged.to_csv('df_final_lyrics.csv')
```
